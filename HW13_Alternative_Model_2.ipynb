```python
import pandas as pd
import numpy as np
from pathlib import Path
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from datetime import date
```

```python
applicant_df_3 = pd.read_csv(
    Path('/Users/ameerirfan/Desktop/FT-BTC-Activities/FT-BTC_Unit_13/applicants_data_2.csv',)
)

applicant_df_3.head()
```

```python
applicant_df_3.dtypes
```

```python
applicant_df_3 = applicant_df_3.drop(columns=['EIN','NAME'])
applicant_df_3.head()
```

```python
cat_variables_3 = list(applicant_df_3.dtypes[applicant_df_3.dtypes=='object'].index)
cat_variables_3
```

```python
enc_3 = OneHotEncoder(sparse=False)

enc_data_3 = enc_3.fit_transform(applicant_df_3[cat_variables_3])
```

```python
enc_df_3 = pd.DataFrame(
    enc_data_3,
    columns=enc_3.get_feature_names(cat_variables_3)
)

enc_df_3.head()
```

```python
non_cat_variables_3 = list(applicant_df_3.dtypes[applicant_df_3.dtypes!='object'].index)
non_cat_variables_3
```

```python
enc_df_3= pd.concat([applicant_df_3[non_cat_variables_3], enc_df_3], axis=1)

enc_df_3.head()
```

```python
Y = enc_df_3['IS_SUCCESSFUL']
Y.head()
```

```python
X = enc_df_3.drop(columns=['IS_SUCCESSFUL'])
X.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y)
```

```python
scaler = StandardScaler()

X_scaler = scaler.fit(X_train)

X_train_scaler_3 = X_scaler.transform(X_train)
X_test_scaler_3 = X_scaler.transform(X_test)
```

```python
num_input_feat_3 = len(X_train.iloc[0])
num_input_feat_3
```

```python
num_output_neurons_3 = 1
```

```python
hidden_nodes_layer1_3 = np.ceil(np.sqrt(num_input_feat * num_output_neurons_3))*3
hidden_nodes_layer1_3
```

```python
hidden_nodes_layer2_3 = np.ceil(np.sqrt(hidden_nodes_layer1_3 * num_output_neurons_3))*3
hidden_nodes_layer2_3
```

```python
hidden_nodes_layer3_3 = np.ceil(np.sqrt(hidden_nodes_layer2_3 * num_output_neurons_3))*3
hidden_nodes_layer3_3
```

```python
nn_3 = Sequential()
```

```python
nn_3.add(
    Dense(
        units=hidden_nodes_layer1_3,
        activation='relu',
        input_dim=num_input_feat
    )
)

nn_3.add(
    Dense(
        units=hidden_nodes_layer2_3,
        activation='relu'
    )
)


nn_3.add(
    Dense(
        units=hidden_nodes_layer3_3,
        activation='relu'
    )
)

nn_3.add(
    Dense(
        units=1,
        activation='sigmoid'
    )
)

nn_3.summary()
```

```python
nn_3.compile(
        loss='binary_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
        
)
```

```python
epochs=100
nn_3.fit(
    X_train_scaler_3,
    Y_train,
    epochs=epochs
)
```

```python
model_loss, model_accuracy = nn_3.evaluate(
                                        X_test_scaler_3,
                                        Y_test,
                                        verbose=2
)

print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
results_df=pd.DataFrame(
                    data={
                        'Date': [date.today()],
                        'Model'   : [2],
                        'Observations':['Adding a hidden layer, increasing epochs'],
                        'Loss'    : [model_loss], 
                        'Accuracy': [model_accuracy],
                        'Hidden layers': 3,
                        'Input Features':[num_input_feat_3],
                        'Output Neurons': [num_output_neurons_3],
                        'Neurons in Hidden Layers1' : [hidden_nodes_layer1_3],
                        'Neurons in Hidden Layers2' : [hidden_nodes_layer2_3],
                        'Epochs'         : [epochs],
                        'Activation Hidden layer 1': ['relu'],
                        'Activation Hidden layer 2': ['relu'],
                        'Activation Output layer 1': ['sigmoid'],
                        'loss':['binary_crossentropy'],
                        'optimizer':['adam'],
                        'metrics':['accuracy'],
                        'Neurons in Hidden Layers3' : [hidden_nodes_layer3_3],
                        'Activation Hidden layer 3': ['relu']
                    })

print(results_df.T)
```
